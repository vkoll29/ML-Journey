### Still using the melbourne datta
### Import both pandas and scikit-learn

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import seaborn as sns
import matplotlib.pyplot as plt

data_path = './melb_data.csv'
melb_data = pd.read_csv(data_path)

melb_data.describe()
```

### 1. Drop all missing values

```python
melb_data = melb_data.dropna(axis=0)
melb_data.describe()
```

### 2. Select the Prediction Target

```python
y = melb_data.Price
y
```

### 3. Select Features

```python
X = melb_data[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
X.describe()
X.head()
```

### 3. First Model

```python
melb_model = DecisionTreeRegressor(random_state=1)
melb_model.fit(X, y)
```

#### 3.1 Make Predictions for the first five houses

```python
print('Actual Values:')
print(y.head().tolist())
print("The Predictions:")
print(melb_model.predict(X.head()))
```

```python
melb_data.head()
```

#### 3.2 Calculate the model's mean absolute error (MAE)

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = melb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

### 4. Validation Data

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
melb_model.fit(train_X, train_y)
val_predictions = melb_model.predict(val_X)
print(melb_model.predict(val_X.head()))
print(val_y.head().tolist())
```

```python
mean_absolute_error(val_y, val_predictions)
```

### 5. Underfitting and Overfitting
##### Use max_leaf_nodes to control fitting

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state = 0)
    model.fit(train_X, train_y)
    y_predicted = model.predict(val_X)
    mae = mean_absolute_error(val_y, y_predicted)
    return mae

for mln in [5, 50, 500, 5000]:
    my_mae = get_mae(mln, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(mln, my_mae))
```

### 5.1 Comparing Different Tree sizes (2)

```python
max_leaf_nodes = [5, 25, 50, 100, 250, 500]
scores = {leaf_size: get_mae(leaf_size,train_X, val_X, train_y, val_y) for leaf_size in max_leaf_nodes}
scores
```
