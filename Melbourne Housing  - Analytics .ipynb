##  Create dataframe from the melbourne housing data csv and preview the first 10 listings

```python
import pandas as pd
data_path = './melb_data.csv'

melb_data = pd.read_csv(data_path)
melb_data.head(10)
```

### 1. View all the regions where houses are located

```python
melb_data.Regionname.unique()
```

### 2. Find all the suburbs and the number of listings each suburb has

```python
melb_data.Suburb.value_counts()
```

### 3. Add the price deviation and variance columns

```python
def deviation_column():
    melb_data['price_deviation'] = (melb_data.Price - melb_data.Price.mean())

    def variance_column():
        melb_data['price_variance'] = (melb_data.price_deviation * melb_data.price_deviation)
        print(melb_data.loc[:10, ['Address', 'Suburb', 'Price', 'price_deviation', 'price_variance']])
    variance_column()
deviation_column()
```

### 4. Find landsize deviation


```python
def landsize_deviation(row):
    row.Landsize = row.Landsize - melb_data.Landsize.mean()
    return row
melb_data.apply(landsize_deviation, axis='columns')
```

### 5. Find the mean landsize of houses in each suburb

```python
melb_data.groupby('Suburb').Landsize.mean()
```

### 6.Get the house where you spend less cash to get more landsize
 You are getting the biggest ratio of landsize to price

```python
bargain_idx = (melb_data.Landsize / melb_data.Price).idxmax()
bargain_home = melb_data.loc[bargain_idx, ['Suburb', 'Rooms', 'Price', 'Landsize']]
bargain_home
```

### 7. Get the bargain home in each suburb

```python
melb_data.groupby('Suburb').apply(lambda df: (melb_data.loc[(df.Landsize / df.Price).idxmax(), ['Address', 'Price', 'Landsize', 'Bedroom2']]))
```

### 8. Get One bedroom houses

```python
def one_br():
    return melb_data.loc[melb_data.Bedroom2 == 1, ['Suburb', 'Address', 'Price', 'Landsize']]
one_br()
```

### 9. The cheapest house in each suburb in each region

```python

melb_data.groupby(['Regionname', 'Suburb']).apply(lambda df: melb_data.loc[df.Price.idxmin(), ['Address', 'Price', 'Bedroom2']])

```

```python
houses_grouped = melb_data.groupby(['Regionname', 'Suburb']).Price.agg([min, max])
houses_grouped = houses_grouped.reset_index()
houses_grouped_ordered = houses_grouped.sort_values(by='min')

houses_grouped_ordered
```
